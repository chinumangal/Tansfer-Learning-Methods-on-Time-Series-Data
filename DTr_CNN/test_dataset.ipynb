```python
#Imports

import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Convert csv-file into panda dataframe
df = pd.read_csv('../dataset/Data/DMG_CMX_600V/AL2007_Bauteil_1/CMX_Alu_Tr_Mat_1_alldata_allforces_MRR_allcurrent.csv')

#to check if everything is correct, print the column names
columns = df.columns.values
print(columns)
```

```python
#oviously not good, only to make sure everything works
df.plot()
```

```python
#Plot axis movement 

figure, ax1 = plt.subplots()
ax1.plot(df[df.columns[0]],df[df.columns[27]],linewidth=0.5,zorder=1, label = "X-Movement")
ax1.plot(df[df.columns[0]],df[df.columns[28]],linewidth=0.5,zorder=1, label = "Y-Movement")
ax1.plot(df[df.columns[0]],df[df.columns[29]],linewidth=0.5,zorder=1, label = "Z-Movement")
ax1.legend()
figure.set_figheight(10)
figure.set_figwidth(10)
```

```python
#plot the velocity and acceleration

figure, ax1 = plt.subplots()
ax1.plot(df[df.columns[0]],df[df.columns[24]],linewidth=0.5,zorder=1, label = "X-Velocity")
ax1.plot(df[df.columns[0]],df[df.columns[25]],linewidth=0.5,zorder=1, label = "Y-Velocity")
ax1.plot(df[df.columns[0]],df[df.columns[26]],linewidth=0.5,zorder=1, label = "Z-Velocity")
ax1.legend()
figure.set_figheight(10)
figure.set_figwidth(10)
```

```python
#Plot only the first x values => to check how long the intervalls need to be

x = 500
df_x = df.head(x)
# Create the plot
figure, ax1 = plt.subplots()

# Plot the cumulative sums
ax1.plot(df_x[df_x.columns[0]],df_x[df_x.columns[29]], linewidth=0.5, zorder=1, label="Z-Movement", color = "green")
ax1.plot(df_x[df_x.columns[0]],df_x[df_x.columns[33]], linewidth=0.5, zorder=1, label="X-Movement", color = "red")

# Create a secondary y-axis
ax2 = ax1.twinx()
ax2.plot(df_x[df_x.columns[0]],df_x[df_x.columns[28]], linewidth=0.5, zorder=1, label="Y-Movement")

# Add the legend
ax1.legend()
ax2.legend()

# Set the figure size
figure.set_figheight(10)
figure.set_figwidth(10)

# Show the plot
plt.show()
```
